```python
import os
import pandas as pd

if __name__ == '__main__':
        pd_all = pd.read_csv("bert_res2/train_results_big2.tsv", sep='\t', header=None)
    data = pd.DataFrame(columns=['polarity'])
    print(pd_all.shape)

    for index in pd_all.index:
        score1 = pd_all.loc[index].values[0]
        score2 = pd_all.loc[index].values[1]
        score3 = pd_all.loc[index].values[2]
        score4 = pd_all.loc[index].values[3]
        score5 = pd_all.loc[index].values[4]
        score6 = pd_all.loc[index].values[5]

        if max(score1, score2, score3, score4, score5, score6) == score1:
            
            data.loc[index+1] = ["1"]
        elif max(score1, score2, score3, score4, score5, score6) == score2:
            
            data.loc[index+1] = ["2"]
        elif max(score1, score2, score3, score4, score5, score6) == score3:
            
            data.loc[index+1] = ["3"]
        elif max(score1, score2, score3, score4, score5, score6) == score4:
            
            data.loc[index+1] = ["4"]
        elif max(score1, score2, score3, score4, score5, score6) == score5:
            
            data.loc[index+1] = ["5"]
        elif max(score1, score2, score3, score4, score5, score6) == score6:
            
            data.loc[index+1] = ["6"]
            
data.to_csv(os.path.join("pre_sample2.csv"))

```
